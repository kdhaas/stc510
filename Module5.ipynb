```python
question_list =[]

for index, rows in df.iterrows():
    iso_questions =[rows.question]
      
    question_list.append(iso_questions)
```

```python
questions_joined = ' /n'.join(str(i) for i in question_list)
```

```python
english_stopwords = set(stopwords.words('english') + list(punctuation) + ['``','/n','\\'])
```

```python
lemmatizer = WordNetLemmatizer()
wordlist = [lemmatizer.lemmatize(x) for x in word_tokenize(questions_joined) if x not in english_stopwords]
```

```python
df['hard_question'] = np.where(df['value'] == '$2000', 1, 0)
df
```

```python
Hard_list =[]

for index, rows in df.iterrows():
    iso_hard =[rows.hard_question]
      
    Hard_list.append(iso_hard)
```

```python
Hard_list
```

```python
list(zip(question_list, Hard_list))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

```python
value_list =[]

for index, rows in df.iterrows():
    iso_value =[rows.value]
      
    value_list.append(iso_value)
```

```python
df2 = pd.DataFrame({'Value':value_list,
                    'hard_question':Hard_list,
                    'question':question_list})
```

```python
X = df2.question.astype(str)
y = df2.hard_question.astype(str)
```

```python
X_Train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
tfidf_vectorizer = TfidfVectorizer(use_idf=True)
X_train_tf = tfidf_vectorizer.fit_transform(X_Train)
X_test_tf = tfidf_vectorizer.transform(X_test)
```

```python
naive_bayes = MultinomialNB()
naive_bayes.fit(X_train_tf, y_train)
predictions = naive_bayes.predict(X_test_tf)
```

```python
print('Accuracy: ', accuracy_score(y_test, predictions))
```
