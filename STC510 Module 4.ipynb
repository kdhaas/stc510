```python
from selenium import webdriver
from time import sleep
import requests
from bs4 import BeautifulSoup
import pandas as pd
CHROMELOC = '/Users/kyle.haas/Desktop/School/STC510/chromedriver'
driver = webdriver.Chrome(executable_path=CHROMELOC)
```

```python
def grabarticle(thisurl):
  thispage = requests.get(thisurl)
  bs = BeautifulSoup(thispage.text,'html.parser')
  description = bs.find(id="postingbody").text[30:-1]
  title = bs.find(class_='postingtitletext').text[1:-1]
  return {'URL':thisurl, 'Description':description,'Title':title}
```

```python
df = pd.read_csv('/Users/kyle.haas/Desktop/School/STC510/cllist.csv')
```

```python
thispage = requests.get('https://phoenix.craigslist.org/search/gms')
bs = BeautifulSoup(thispage.text,'html.parser')
latest = bs.find(id='search-results')
```

```python
links = latest.find_all(class_='result-title')
for eachlink in links:
  full_url = eachlink['href']
  if not df.URL.str.contains(full_url).any():
    print('getting',full_url)
    post = grabarticle(full_url)
    sleep(20)
    df = df.append(post,ignore_index=True)    
```
